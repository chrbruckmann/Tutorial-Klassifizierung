```python
# Mit dem gelernten Netz kann man im Prinzip alles klassifizieren, 

from bing_image_downloader.downloader import download
# !pip install bing-image-downloader  # falls modul nicht installiert
```

```python
query_list = ['katzen', 'hunde'] 
pic_num = 200 # Anzahl der Bilder pro Begriff


for query_string in query_list:
    download(query_string, limit=pic_num,  output_dir='download', adult_filter_off=False, force_replace=False, timeout=60, verbose=True)
```
